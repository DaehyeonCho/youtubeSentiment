```python
import pandas as pd

# 합칠 CSV 파일들의 경로를 리스트로 저장
files = ['results1.csv', 'results2.csv', 'results3.csv', 'results4.csv', 'results5.csv', 'results6.csv', 'results7.csv', 'results8.csv', 'results9.csv', 'results10.csv']

# 빈 DataFrame 생성
result = pd.DataFrame()

# CSV 파일을 순회하면서 데이터를 읽어와 합침
for file in files:
    df = pd.read_csv(file)
    result = pd.concat([result, df], ignore_index=True)

# 결과를 새로운 CSV 파일로 저장
result.to_csv('merged_result.csv', index=False)
```
